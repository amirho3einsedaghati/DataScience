```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/presidents.csv") 
df
```

```python
df.columns = [item.lower().strip() for item in df.columns]
df.head()
```

solution 1 :

```python
df["first"] = df["president"]
df["first"] = df["first"].replace('[ ].*', '', regex= True)

df["last"] = df["president"].replace('(.*)(?=\s.*)', '', regex= True)

df.head()
```

```python
del df["first"]
del df["last"]
```

solution 2 :

```python
df["first"] = df["president"]
df["first"] = df["first"].replace("\s.+", "", regex= True)

df["last"] = df["president"].replace("([\w]+)(?=\s[.]*)", "", regex= True)

df.head()
```

```python
df.drop("first", inplace= True, axis= 1)
df.drop("last", inplace= True, axis= "columns")
```

solution 3 :

```python
def splitName(row):
    row["first"] = row["president"].split(" ")[0]
    row["last"] = row["president"].split(" ")[-1]
    
    return row

df = df.apply(splitName, axis= 1)
df.head()
```

solution 4:

```python
del df["first"]
del df["last"]
```

```python
pattern = "(^.*)(?:\s)(.*$)"

df["president"].str.extract(pattern)
```

```python
pattern = "(?P<FirstName>^[\w]*)(?:.* )(?P<LastName>[\w]*$)"

names_df = df["president"].str.extract(pattern)
names_df.head(10)
```

```python
df["first"] = names_df["FirstName"]
df["last"] = names_df['LastName']
df.head()
```

```python
df["born"] = df["born"].str.extract('([\w]{3}\s\d{1,2},\s[\w]{4})')
df["born"].head()
```

```python
df["born"] = pd.to_datetime(df["born"])
df["born"].head()
```

```python
df.head()
```

```python
df.set_index("#", inplace= True)
```

```python
df = df.rename(columns= {"age atstart of presidency" : "age at start of presidency",
                   "post-presidencytimespan" : "post-presidency timespan"})
df.head()
```

```python
die_df1 = df.loc[1:38, "died"]
die_df2 = df.loc[40:41, "died"]
die_df = die_df1.append(die_df2)
die_df = pd.to_datetime(die_df)
die_df
```

```python
living_df = df[df["died"] == "(living)"]
living_df["died"]
```

```python
df["died"] = die_df.append(living_df['died'])
df
```

```python
df["died"]
```

```python
df["Age at start of presidency"] = df["age at start of presidency"].str.extract("(.*)(?=[A-Z])")
df["the start year of presidency"] = df["age at start of presidency"].str.extract("(?:.*)([A-Z].*)")

df["the start year of presidency"] = pd.to_datetime(df["the start year of presidency"])    
```

```python
df["age atend of presidency"].replace(df.loc[9, "age atend of presidency"], "68 years, 54 daysApr 4, 1841[b]", inplace= True)
df["age atend of presidency"] = df["age atend of presidency"].str.extract("(.*[\w]{3}\s\d{1,2}\,\s\d{4})")
df["age atend of presidency"]
df.head(10)
```

```python
df["Age atend of presidency"] = df["age atend of presidency"].str.extract("(.*)(?=[A-Z])")
df["the end year of presidency"] = df["age atend of presidency"].str.extract("(?:.*)([A-Z].*)")

df["the end year of presidency"] = pd.to_datetime(df["the end year of presidency"]) 
df["the end year of presidency"].head(10)
```

```python
del df["president"]
del df["age at start of presidency"]
del df["age atend of presidency"]

df.rename(columns= {"Age at start of presidency" : "age at start of presidency",
                    "Age atend of presidency" : "age atend of presidency",
                   "first" : "president first name",
                   "last" : "president last name"}, inplace= True)


df.head()
```
